#Books SQL

##Import sqlite3 and Connect

```python
import sqlite3

connection = sqlite3.connect('books.db')

```

##Viewing the Table of Contents

###Import pandas, set column limits, and read SQL

```python
import pandas as pd

pd.options.display.max_columns = 10

pd.read_sql('SELECT * FROM authors', connection, index_col=['id'])
```

##Show Titles

```python
pd.read_sql('SELECT * FROM titles', connection)
```

##Show author ISBN

```python
pd.read_sql('SELECT * FROM author_ISBN', connection)
```

##1 - SELECT

```python
pd.read_sql('SELECT first, last FROM authors', connection)
```

##2- WHERE

```python
pd.read_sql("""SELECT title, edition, copyright
            FROM titles
            WHERE copyright > '2016'""", connection)
```

###LIKE and pattern matching zero or more

```python
pd.read_sql("""SELECT id, first, last
            FROM authors
            WHERE last LIKE 'D%'""",
            connection, index_col=['id'])
```

###pattern matching any character

```python
pd.read_sql("""SELECT id, first, last
            FROM authors
            WHERE first LIKE '_b%'""",
            connection, index_col=['id'])
```

##3 - Order By

```python
pd.read_sql('SELECT title FROM titles ORDER BY title ASC',
            connection,)
```

###Sorting by multiple columns

```python
pd.read_sql("""SELECT id, first, last
            FROM authors
            ORDER BY last, first""",
            connection, index_col=['id'])
```

```python
pd.read_sql("""SELECT id, first, last
            FROM authors
            ORDER BY last DESC, first ASC""",
            connection, index_col=['id'])
```

###Combine WHERE and ORDER BY

```python
pd.read_sql("""SELECT isbn, title, edition, copyright
            FROM titles
            WHERE title LIKE '%How to Program'
            ORDER BY title""", connection)
```

##4 - INNER JOIN
###Merging Data from multiple tables


```python
pd.read_sql("""SELECT first, last, isbn
            FROM authors
            INNER JOIN author_ISBN 
                ON authors.id = author_ISBN.id
            ORDER BY last, first""", connection).head()
```

##5 - INSERT INTO
###using cursor object

```python
cursor = connection.cursor()

cursor.execute("""INSERT INTO authors (first, last)
                    VALUES ('Sue', 'Red')""" )
```

```python
pd.read_sql('SELECT id, first, last FROM authors' , 
            connection, index_col=['id'])
```

##6 - UPDATE


```python
cursor = cursor.execute("""UPDATE authors SET last='Black'
                            WHERE last='Red' and first='Sue'""" )
```

```python
cursor.rowcount
```

```python
pd.read_sql('SELECT id, first, last FROM authors' ,
            connection, index_col=['id'])
```

##7 - DELETE FROM

```python
cursor = cursor.execute('DELETE FROM authors WHERE id=6')

cursor.rowcount
```
